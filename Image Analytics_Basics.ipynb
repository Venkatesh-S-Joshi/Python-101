```python

sales = float(input("Enter the quarterly sales figures: ₹"))

# Prompt the user to input the total attendance days for the quarter
attendance = int(input("Enter the total attendance days for the quarter: "))

# Prompt the user to input the average customer satisfaction rating
customer_satisfaction = float(input("Enter the average customer satisfaction rating (1.0 to 5.0): "))

# Prompt the user to input the project completion rate as a percentage
project_completion_rate = float(input("Enter the project completion rate (as a percentage): "))

# Determine the performance category based on sales, attendance, customer satisfaction, and project completion rate
if (sales >= 300000 and attendance >= 66 and 
    customer_satisfaction >= 4.5 and project_completion_rate == 100):
    print("Performance Category: Outstanding")
elif ((200000 <= sales < 300000) or (54 <= attendance < 66) or 
      (4.0 <= customer_satisfaction < 4.5) or (80 <= project_completion_rate < 100)):
    print("Performance Category: Excellent")
elif ((150000 <= sales < 200000) or (45 <= attendance < 54) or 
      (3.5 <= customer_satisfaction < 4.0) or (60 <= project_completion_rate < 80)):
    print("Performance Category: Good")
elif ((75000 <= sales < 150000) or (30 <= attendance < 45) or 
      (3.0 <= customer_satisfaction < 3.5) or (40 <= project_completion_rate < 60)):
    print("Performance Category: Average")
elif (sales < 75000 or attendance < 30 or 
      customer_satisfaction < 3.0 or project_completion_rate < 40):
    print("Performance Category: Below Average")
else:
    print("Invalid input.")
```

# Image Analytics

## Read, Show and Write images

```python
import cv2
img = cv2.imread('BTC3.jpg')
cv2.imshow('show',img)
cv2.imwrite('photo.jpg',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Colour to Grey scale

```python
import cv2
image = cv2.imread('BTC3.jpg')
greyImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray_image.png', greyImage)
cv2.imshow('Color_image',image)
cv2.imshow('Grey_image', greyImage) 
cv2.waitKey(20000)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread("BTC3.jpg")
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("GrayImage.jpg",grayImg)
cv2.imshow("Original",img)
cv2.imshow("GrayImage",grayImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Image Resize

```python
import cv2
import imutils
import os

# Check and print the current working directory
print("Current working directory:", os.getcwd())

# Load the image (provide the correct full path if necessary)
img = cv2.imread("BTC3.jpg")

# Check if the image was loaded successfully
if img is None:
    print("Error: Could not load image. Check the file path or file name.")
else:
    # Resize the image to a width of 500 pixels
    resizedimg = imutils.resize(img, width=500)

    # Save the resized image
    cv2.imwrite('resizedImage.jpg', resizedimg)

    print("Resized image saved as 'resizedImage.png'")
```

```python
import cv2
import imutils
img = cv2.imread('BTC3.jpg')
resizedImg = imutils.resize(img, width=500)
cv2.imwrite('resizedImage.jpg', resizedImg)
```

## Gausian Blur

```python
import cv2
img = cv2.imread('BTC3.jpg')
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gaussianImg = cv2.GaussianBlur(grayImg, (21, 21), 0)
cv2.imwrite("GaussianBlur.jpg", gaussianImg)
```

## Threshold

```python
import cv2
img=cv2.imread("BTC3.jpg")
grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gaussBlur = cv2.GaussianBlur(grayImg,(21,21),0)
thresholdImg = cv2.threshold(grayImg,150,255,cv2.THRESH_BINARY)[1]
cv2.imwrite("threshold.jpg",thresholdImg)
```

## drawing rectangle

```python
import cv2
imgrect = cv2.rectangle(img, (100, 50), (128,80), (0, 255, 0), 2)
cv2.imshow('rect',imgrect)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
